```python
%matplotlib inline
```

```python
# Import matplotlib library
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# set the y-axids to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create the plot and add label for legend
plt.plot(x_axis, y_axis, label='Boston')

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# set the y limit between 0 and 45
plt.ylim(0,45)

# create a title
plt.title("PyBer Fare by Month")

# add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2,label="Boston")

# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the bar plot.
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color = "green", label = "Boston")

# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create title
plt.title("PyBer Fare by Month")

# Add legend.
plt.legend()
```

```python
# Create horizontal bar plot
plt.barh(x_axis, y_axis)
```

```python
# Get months in ascending order gca = "get current axis"
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()

# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create title
plt.title("PyBer Fare by Month")

# Add legend.
plt.legend()
```

```python
# Create vertical bar with ax.plt()

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create horizontal bar with ax.plt()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label="Chicago")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Bubble plot with with size variable
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Increase the size of the data points
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension to replace for loop
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# Create scatter plot using object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create bubble chart using object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create pie chart using MATLAB method
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange",
          "gold","indianred","tomato","mistyrose"]

# Increase the size of the pie chart
plt.subplots(figsize=(8,8))

# Explode a pie wedge (July)
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)

# Add percentage of each wedge on pie chart
plt.pie(y_axis, 
        labels = x_axis, 
        explode = explode_values, 
        colors = colors,
        autopct = '%.1f%%')

# Display pie chart
plt.show()
```
